```python
# Suppress all warnings
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Path to the CSV file
file_path = "C:\\Users\\bashi\\OneDrive\\Desktop\\plan of action\\Data\\csv files\\number-of-immigrants-in-MB-2000-2022.csv"

# Load the dataset from the local path
df = pd.read_csv(file_path)

# Preprocess the 'Number of Immigrants' column
df['Number of Immigrants'] = df['Number of Immigrants'].str.replace(',', '').astype(int)

# Prepare features and target
X = df[['Year']]
y = df['Number of Immigrants']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate predictions using the linear regression model
years_to_predict = np.arange(2001, 2031).reshape(-1, 1)
predicted_immigrants = model.predict(years_to_predict)

# Create a visual plot with enhanced styles
plt.figure(figsize=(10, 6))
plt.stackplot(df['Year'], df['Number of Immigrants'], alpha=0.4, colors=['lightblue'], edgecolor='darkblue')
plt.plot(years_to_predict, predicted_immigrants, label='Trend Line', color='green', linestyle='dashed', linewidth=2, alpha=0.7)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Immigrants', fontsize=12)
plt.title('Number of Landed Immigrants to Manitoba', fontsize=12)
plt.legend(loc='lower right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Remove figure edges
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set 1: Manitoba Units
unit_types = ["Bachelor", "1 Bedroom", "2 Bedroom", "3+ Bedroom"]
manitoba_units = [4626, 5110, 3744, 6577]

# Set 2: Winnipeg Units
winnipeg_units = [3081, 2678, 2677, 3764]

# Set 3: Ownership Types in Winnipeg
ownership_labels = ["Government", "Non-Profit", "Coop"]
ownership_types = [8904, 2627, 602]

# Set 4: Construction Year in Winnipeg
construction_labels = ["Pre 1980", "1980 - 1986", "1987 - 1995", "1996 - 2003", "2004 - 2019"]
construction_year = [8975, 1123, 1355, 0, 748]

# Create a 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Determine the common y-axis limit
ylims = (0, max(manitoba_units + winnipeg_units + ownership_types + construction_year) * 1.1)

# Plot 1: Manitoba Units
sns.barplot(x=unit_types, y=manitoba_units, palette='viridis', ax=axes[0, 0])
axes[0, 0].set_title('Units in Manitoba', fontsize=14)
axes[0, 0].set_ylim(ylims)

# Plot 2: Winnipeg Units
sns.barplot(x=unit_types, y=winnipeg_units, palette='viridis', ax=axes[0, 1])
axes[0, 1].set_title('Units in Winnipeg', fontsize=14)
axes[0, 1].set_ylim(ylims)

# Plot 3: Ownership Types in Winnipeg
sns.barplot(x=ownership_labels, y=ownership_types, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title('Ownership Types in Winnipeg', fontsize=14)
axes[1, 0].set_ylim(ylims)

# Plot 4: Construction Year in Winnipeg
sns.barplot(x=construction_labels, y=construction_year, palette='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Construction Year in Winnipeg', fontsize=14)
axes[1, 1].set_ylim(ylims)

plt.tight_layout()
# plt.savefig('social and affordable units.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
conditions = ['Excellent', 'Good', 'Average', 'Fair']
values = [290, 427, 447, 11031]

# Define base colors
colors = ['darkgreen', 'limegreen', 'gray', 'yellow']

# Generate gradient colors by blending each base color with white
gradient_colors = []
for color in colors:
    gradient_colors.append([plt.cm.colors.to_rgba(color, alpha) for alpha in np.linspace(0.3, 1, 10)])

# Create custom colormap
from matplotlib.colors import LinearSegmentedColormap
colormaps = [LinearSegmentedColormap.from_list("custom_cmap", colors) for colors in gradient_colors]

# Pie chart
fig, ax = plt.subplots(figsize=[10,6])
wedges, texts, autotexts = ax.pie(values, labels=conditions, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12}, pctdistance=0.85, colors=[cmap(0.6) for cmap in colormaps])

# Add gradient effect
for wedge, cmap in zip(wedges, colormaps):
    wedge.set_edgecolor("white")
    wedge.set_facecolor(cmap(0.6))

plt.axis('equal')
# plt.savefig('conditions.png')
plt.show()
```

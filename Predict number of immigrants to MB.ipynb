```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset from the provided link
dataset_url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQGYtXFqTUf07GIyEsQzMvdbzZafHpUZ5GPacWTii5eOwi_ZwU7EX2E4_rleuhnNtRrPHZJ0wVA1hQZ/pub?output=csv"
df = pd.read_csv(dataset_url)

# Preprocess the 'Number of Immigrants' column
df['Number of Immigrants'] = df['Number of Immigrants'].str.replace(',', '').astype(int)

# Prepare features and target
X = df[['Year']]
y = df['Number of Immigrants']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate predictions using the linear regression model
y_pred = model.predict(X)

# Create a visual plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Number of Immigrants'], label='Actual Data', color='blue')
plt.plot(df['Year'], y_pred, label='Trend Line', color='orange')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Number of Immigrants to Manitoba (2001-2022) with Trend Line')
plt.legend()
plt.grid(True)
plt.show()
```
